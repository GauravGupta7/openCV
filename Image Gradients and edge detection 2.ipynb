### SobelX and SobelY
Here we are trying edge detection using the **SobelX and SobelY** method. 
* When we apply SobelX method, the change in direction in the intensity of color is along the X axis
* When we apply SobelY method, the change in direction in the intensity of color is along the Y axis

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("data/data_cv2/sudoku.png", cv2.IMREAD_GRAYSCALE)
SobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0)
SobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)

SobelX = np.uint8(np.absolute(SobelX))
SobelY = np.uint8(np.absolute(SobelY))

SobelCombined = cv2.bitwise_or(SobelX, SobelY)

titles = ['image', 'SobelX', 'SobelY', 'Sobel Combined']
images = [img, SobelX, SobelY, SobelCombined]
for i in range(4):
    plt.subplot(2, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
