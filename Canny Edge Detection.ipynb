### Introduction
The Canny Edge Detection algorithm used to detect wide range of edges in images. It involves the following steps:
* Noise Reduction
* Gradient Calculation
* Non-Maximum Suppression
* Double Threshold
* Edge Tracking by Hysteresis  

**IMP:** Canny Edge Detection algo is based on gray scale images and so the Image needs to be converted to gray scale before applying the algorithm. 

**Canny Edge Detection method is one of the best method of edge detection**. It has very little of noise in it.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('data/data_cv2/lena.jpg', 0)
canny = cv2.Canny(img, 100, 200)

titles = ['Original', 'Canny']
images = [img, canny]
for i in range(2):
    plt.subplot(1, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
