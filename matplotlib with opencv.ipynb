```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('data/data_cv2/lena.jpg')

#opening the image using openCV
cv2.imshow('Image', img)

#opening the image using matplotlib
plt.imshow(img)
plt.show()

"""The difference in the color scheme of the images is because openCV reads in the BGR scheme while matplotlib 
   reads in the RBG format."""
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Using the color converter
#opening the image using openCV
cv2.imshow('Image', img)

#converting the color scheme
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#opening the image using matplotlib
plt.imshow(img)
plt.xticks([]), plt.yticks([])      #to remove axes from the matplotlib image
plt.show()

"""The difference in the color scheme of the images is because openCV reads in the BGR scheme while matplotlib 
   reads in the RBG format."""
cv2.waitKey(0)
cv2.destroyAllWindows()
```
