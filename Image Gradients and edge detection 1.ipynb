### Image Gradients and Edge Detection
* Image Gradient is a directional change in the intensity or color of an image
Gradient descent in an important concept necessary for edge detection in image processing
Here we will work on Edge Detection using gradient decend and the **Laplacian Method**


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("data/data_cv2/messi5.jpg", cv2.IMREAD_GRAYSCALE)
lap = cv2.Laplacian(img, cv2.CV_64F)
lap = np.uint8(np.absolute(lap))

titles = ['image', 'Edges']
images = [img, lap]
for i in range(2):
    plt.subplot(1, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
