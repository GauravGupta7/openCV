## Histograms

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("data/data_cv2/lena.jpg")

# Splitting the image into component colors
b, g, r = cv2.split(img)

cv2.imshow("Img", img)
cv2.imshow("b", b)
cv2.imshow("g", g)
cv2.imshow("r", r)

# Histograms using Matplotlib
plt.hist(b.ravel(), 256, [0,256])
plt.hist(g.ravel(), 256, [0,256])
plt.hist(r.ravel(), 256, [0,256])
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread("data/data_cv2/lena.jpg")
hist = cv2.calcHist([img1], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```
